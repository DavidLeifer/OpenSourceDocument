  Vocabulary Review

  - *Descriptive Statistics* - summarizing data using quantitative methods.
  - *Parametric Statistics* - makes assumptions about the data
    - the data is normally distrubted around the mean, or symmetrical like the typical bell curve graph
    - at minimum requires 30 data points. Most studies use 100 or more to have good results.
      - Do not use significant or confidence to describe the results since those are statistical tests
  - *Types of Data*
    - *Ordinal* - it is categorical with clear differences in rankings
      - the pain scale CSV that I made up for this class is an example
    - *Nominal* - there is no order between the value
      - W3 Schools Python example: blonde, brunette, etc.
    - *Interval* - or numerical data that is equally spaced integers
      - for example income values: $500, $1000, $1500, etc.
      - calories and alcohol columns in B0.csv are examples
        - calories are the actual measureable amount of food energy consumed
        - alcohol is the number of alcohol servings consumed
  - There are several statistical errors treating ordinal data as if it were parametric since they aren't real world measurements
  - The mean is typically used in parametric statistics
  - The median is typically used in non-paramteric statistics
  - More information:
  https://stats.oarc.ucla.edu/other/mult-pkg/whatstat/what-is-the-difference-between-categorical-ordinal-and-interval-variables/

  Parametric Vocabulary Review

  - *Mean* - Greek letter mu - the sum of all numbers divided by the count of values. Used in parametric statistics.

  - *Median* - the middle most number of the dataset and is typically used in non-paramteric statistics

  - *Standard deviation* - each value in the dataset minus the mean. That number is then squared. Then sum each of those numbers and divide by the count of numbers. The square root of that number is then the standard deviation.

    - Represented by the Greek lower case letter for sigma.

    - Also the second central moment of the distribution.
    
      - A statistician can calculate the skewness and kurtosis which represent the third and forth moment.

      - An obssesive staistician can calculate 7 or more moments but they are more influenced by the outliers.

    - It is interesting that the Google Sheets functions for mean, count, and standard deviation return different values than the Python implementation. There are two possible causes:
      1. Python is using feature reduction to reduce hardware use on repetitive numbers, as examined in the previous class.
      2. Asymptotic theory with infinite vs finite data samples

  **Central Moment Vocabulary**

  Recall that variance or standard deviation is the second central moment of a probablility distribution and is represented by lower case sigma.

  - Skewness - is the third central moment. There are several version of this equation. The one we will use is defined here: subtracts each value in the list by the overall mean and multiplies itself three times or to the third power. The sum of that number is then calcualted. The sum is divided by the count of values in the list. That number is divided by the standard deviation to the power of 3 (or cubed).

    - The equation is for the entire population of the data or all the numbers.

    - In contrast to a sample of the dataset, which subtracts 1 from the count (N-1) before the first division.

    - There was a long argument about whether skewness measures peakedness or flatness of the bell curve and can be read in full on the NIH here:

      - https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4321753/

    - The equation is defined as such by the NIST and in the International Encyclopedia of Statistical Science (PDF download from the University of Texas):

        - https://www.itl.nist.gov/div898/handbook/eda/section3/eda35b.htm

        - https://lbj.utexas.edu/sites/default/files/file/news/Skew.pdf

  - Kurtosis - UCLA provides definition for three kurtosis equations:

    - Two sums of powered deviation scores, which are used later:

      - s2 = (xi - mu) ^ 2

      - s4 = (xi - mu) ^ 4

    - Calculate the standard deviation:

      -  m2 = s2/n

      -  m4 = s4/n

    - Above and beyond unbiased estimate

      - V(x) = s2/n-1

    - The first defintion of kurtosis used by SAS summarize function and cites a textbook:

      - kurtosis = (m4/m2^2) - 3

      - Snedecor, G.W. and Cochran, W.G. (1967) Statistical Methods, Sixth Edition. Ames, Iowa: Iowa State University Press.
        
    - The second definition is the same formula as skewness but replace the 3 with 4. It is used in the STATA program and cites a different textbook.

      - kurt = m4 / m2

      - Bock, R.D. (1975) Multivariate Statistical Methods in Behavioral Research. New York: McGraw-Hill.

    -  The final formula is overkill for this course but is worth mentioning if you are bored and out of money for beer or weed. It is used in SPSS and cites a more recent textbook. It is noted as being less biased except for non-normal samples or data that is not symetrically bell shaped.

      - Sheskin, D.J. (2000) Handbook of Parametric and Nonparametric Statistical

      - Example is calories and alcohol consumption in the B0.csv.

  - Covariance

      - The first step is to calculate the covariance, which is the numerator of the formula.

      - Calculate the first value xi minus the x mean.

      - Calculate the second value yi minus the y mean.

          - Multiply these two value together.

      - The result is the covariance, which is an unscaled mathematical relationship.
