>- The <i>csv_flipper</i> function is present in both methods of bad code:

```python
def csv_flipper(csv_list, col_num):
  '''
    1-5 would by flipped to 5-1 making the
    identifier column less useful. Since 'ID' is the first
    value in every row, if the count ('i') is divisible by
    the length of the row ('col_num') it is not flipped.
  '''
  i = 0
  csv_flipped = []
  for j in csv_list:
    # Here is the unflipped 'ID' value.
    if i % col_num == 0:
      csv_flipped.append(j)
    elif j == str(5):
      n = str(1)
      csv_flipped.append(n)
    elif j == str(4):
      n = str(2)
      csv_flipped.append(n)
    elif j == str(3):
      n = str(3)
      csv_flipped.append(n)
    elif j == str(2):
      n = str(4)
      csv_flipped.append(n)
    elif j == str(1):
      n = str(5)
      csv_flipped.append(n)
    else:
      csv_flipped.append(j)
    i += 1
  return csv_flipped
```

>- The dataset has been traversed one time.

>- The values for stamina were flipped since they did not meet the criteria to be excluded in <i>csv_flipper</i> and are now opposite the intended meaning.

>- Instead another function extracts the column from the original <i>csv_list</i> by making a list of the index for the 4th value in every row.

```python
def stm_extract(csv_list, col_num):
  i = col_num - 1
  stamina_col = []
  for j in csv_list[3:]:
    if i == col_num - 1:
      stamina_col.append(j)
      i = 0
    else:
      i += 1
  return stamina_col
```

The second part of the program has two options to return the flipped list without the stamina column.

<i>Method 0</i> :

>- The header and first four values are skipped in the loop's first conditional.

>- The second conditional skips appending values from stamina onto <i>flip_list_rm</i> by keeping track of the fourth value in every row with <i>col_num</i>.

```python
def stm_remove(flip_list, col_num):
  flip_list_rm = []
  og_col_num = col_num
  for i in range(len(flip_list)):
    if i == 3:
        col_num = i + col_num
    elif i == col_num:
      col_num += og_col_num
    else:
      flip_list_rm.append(flip_list[i])
  return flip_list_rm
```

<i>Method 1</i> : The original list is preserved and uses the built-in pop function to remove the values from the list.

```python
def stm_remove(flip_list, col_num):
  og_col_num = col_num - 1
  og_flip_list_len = len(flip_list)
  col_len = 32
  for i in range(len(flip_list)):
    if i ==(og_flip_list_len - (col_len)):
      break
    elif i == 3:
      flip_list.pop(i)
      col_num = i + col_num - 1
    elif i == col_num:
      flip_list.pop(i)
      col_num += og_col_num
  return flip_list
```
