**Homework Notes**

>- The next few weeks are spent learning how to display XYZ tiles and building the web application with 'npm'.

>- There are several additional methods in the OpenLayers documentation that can be experimented with for added functionality.

>- Imports to the JavaScript code include references to the methods found in the <i>Map</i> file, which creates the <i>map</i> div.

```python
import Map from 'ol/Map';
```

>- The next <i>View</i> import controls the map's center, zoom, and projection parameters.

```python
import View from 'ol/View';
```

>- The final references include setting links to  XYZ tile methods with the imports <i>TileLayer</i> and <i>XYZ</i>.

```python
import TileLayer from 'ol/layer/Tile';
import XYZ from 'ol/source/XYZ';
```

>- The methods look like the code below. The URL is pointing to the standard OpenStreetMap tile server.

```python
new Map({
  target: 'map',
  layers: [
    new TileLayer({
      source: new XYZ({
        url: 'https://tile.openstreetmap.org/{z}/{x}/{y}.png'
      })
    })
  ],
  view: new View({
    center: [0, 0],
    zoom: 2
  })
});
```
