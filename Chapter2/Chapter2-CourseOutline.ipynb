2. Managing User Permissions

  **Week 1 Homework Reminder**

  - Permission for a file include who in can read, write, or execute a file or directory.
  
  - Most retail devices connected to the internet can be logged into from other computers using the ssh (secure shell) protocol.

    - If someone accesses your computer, they could install a virus or steal other information.

    - Some hacking programs on the internet are automated and have no owner, making it impossible for retribution or to stop since the program jumps from device to device similar to the flu virus.

  - Hopefully this course will shed light on not falling for automated nonsense.

  **Homework Notes**

  - Below is not an exhaustive list of permission related commands and includes only the basics. Read the documentation if you have more questions.

  - List permissions of a file or directory:
    ```
    ls-l
    ```

  - Allow the user of a file to read and write:
    ```
    chmod u+rw /some/file.py
    ```
  - The first letter of the command is the user or group:
    ```
    User                                      Letter

    The user who owns it                      u
    Other users in the file's group           g
    Other users not in the file's group       o
    All users                                 a
    ```
  - The '+' or '-' sign adds or subtracts a permission. Command line permission list can be a letter or number:
    ```
    Command Explanation                       Letter    Number

    Read                                      r         4
    Write                                     w         2
    Execute (or access for directories)       x         1
    ```
  - Command to change the ownership of all the files and directories in a directory where -R is a flag and sudo temporarily gives root access to the user without requiring the root password:
    ```
    sudo chown -R usrname: /this/dir/
    ```
  - Other flags can be added to modify the command:

    ```
    Apply to all elements in directory        R
    Set user or group ID on execution         s
    Restricted deletion flag                  t
    The permissions for a user                u
    The permissions for other users           g
    Permissions for users not in the group    o
    ```

  - Tomcat9 GUI Configuration

  - To use the GUI, open up “tomcat-users.xml” with this command:
    ```
      sudo nano /var/lib/tomcat8/conf/tomcat-users.xml

      <role rolename="manager-gui"/>
      <role rolename="admin-gui"/>

    ```
  - Save the file and restart tomcat8:
    ```
      sudo systemctl restart tomcat8
    ```
  - Now you can point the browser to the localhost at port 8080:

  http://localhost:8080/

  or

  http://127.0.0.1:8080/


  - This is running only on your computer and is not connected to the internet.
  
  - The IP address would be the address of the server or the domain name the IP address redirects.

  - http: Hyper Text Transfer Protocol

    - No encryption. The server can access your device.

  - https: Hyper Text Transfer Protocol Secure

    - Encrypts the communication between the browser and the server and uses a certificate system to ensure text packages are not modified.

9. Installing Geoserver: Part 1 of 2

  **Week 3 Homework Due**

  **Week 9 Homework: Geoserver**

    Grading Scale:
    0. Did not attempt.
    1. Installed Geoserver

    - Navigate to the “webapps” folder within your “tomcat” installation with this command:
    ```
    cd /var/lib/tomcat9/webapps
    ```
    - Next, download the latest Geoserver zip from their website:

    ```
      sudo wget <url_path_to/geoserver-version.x-latest-bin.zip>
    ```

    - Install the “unzip” utility if it is not already installed:

    ```
      sudo apt-get install unzip
    ```

    - Move the zip to the webapps directory and unzip the package:

    ```
      sudo mv /User/Downloads/geoserver-version.x-latest-bin.zip /var/lib/tomcat9/webapps/
      sudo unzip /var/lib/tomcat9/webapps/geoserver-version.x-latest-bin.zip
    ```

    - Then make the .war executable for the current user by entering:

    ```
      sudo chmod u+x /var/lib/tomcat8/webapps/geoserver.war
    ```

    - Start Geoserver:

    ```
      cd /var/lib/tomcat9/webapps/geoserver/bin/
      sh startup.sh
    ```

    - Finally, restart “tomcat8”:

    ```
      sudo service tomcat8 restart
    ```

    - Point your browser at the following address:
    http://localhost:8080/geoserver/web/
    
    - The default behavior of Geoserver is to set user to “admin” and password to “geoserver” but you can change it either in the web application or in the appropriate configuration XML.

  **Week 13 Homework: Configuring Geoserver**

    Grading Scale:
    0. Did not attempt.
    1. Configured Geoserver

  - Configure the Geoserver installation to allow the upload of a timeseries of a raster of your choice using the Image Mosaic plugin.
  
  - Example rasters can be downloaded from the USGS Earth Explorer web application. You may need to use QGIS, ArcMap, or Python to clip the tiff file size smaller or downsample the image.

  - After downloading the data onto the Linux OS, change permissions on the folder to upload onto Geoserver. Navigate to the “data” folder and make a directory to host our files:

  ```
    cd /var/lib/tomcat9/webapps/geoserver/data/data
    sudo mkdir timeseries
    cd timeseries
  ```

  - The two .properties files look like this:

  ```
    timeregex.properties
    regex=[0-9]{8}
    indexer.properties
    TimeAttribute=ingestion
    ElevationAttribute=elevation
    Schema=*the_geom:Polygon,location:String,ingestion:java.util.Date,elevation:Integer
    PropertyCollectors=TimestampFileNameExtractorSPI[timeregex](ingestion)
  ```

  - Change permissions on the last /data/ folder:

    ```
      sudo chmod -R u+rw /var/lib/tomcat9/webapps/geoserver/data/data/timeseries
    ```

  - Go to the web application in your browser.

  - Under “Data” on the left hand side of the screen go to “Stores”.

  - Select “Add new stores”.
  - Select “ImageMosaic” and “Browse”. Navigate to the folder you put the data in (data/timeseries)
  - Select “OK”.
  
  - Fill out “Data Source Name” and “Description” then save.

  - Click “Publish” on the next screen.
  - Click “Dimensions” and enable “Time”. “Presentation”
  should be set to “List”.
  - Then select “Save”.

  If it worked, you can go to “Layer Preview” and click “timeseries” and view the tiffs as Web Map Services (WMS).

  **Notes**

  - ESRI (Environmental Systems Research Institute) is a developer for GIS software such as ArcMap and ArcPro and commands most of the market share by licensing to governments and private companies.

  - OpenLayers is an open source web development library with numerous features that are similar to the ESRI's proprietary ArcGIS for JavaScript library.

    - Open source software has almost no licensing or warranties about mathematical accuracy and reliability. These are usually free and are written by hobbyists and educational instructors that enjoy programming.

    - Proprietary software is developed by private or public organizations that charge money to ensure their products have a standard of accuracy and reliability.

  - This week a website will be developed to view the WMS as a time slider in a browser.

 - Modify HTML/CSS/JS to make a Time Slider by navigating to the www folder:

    ```
      cd /var/lib/tomcat8/webapps/geoserver/data/www
    ```

  - Create a file name rasterTime.html

    ```
      sudo nano rasterTime.html
    ```

  - The HTML file references two files from OpenLayers that give access to numerous CSS and JavaScript (JS) methods. Copy this markdown code into the file:

    ```
    <!DOCTYPE html>
    <html>
    <head>
    <title>WMS Time</title>
    <link rel="stylesheet" href="https://openlayers.org/en/v4.6.5/css/ol.css" type="text/css">
    <script src="https://openlayers.org/en/v4.6.5/build/ol.js"></script>
    ```
      
    
  - CSS is added for styling the HTML. This can be included as a seperate .CSS file with markdown similar to the OpenLayers references:

    ```
    <style>
    /* Slider styling */
    .slidecontainer {
    width: 120px;
    margin-top: -30px;
    margin-left: 10px;
    z-index: 100px;
    }
    .slider {
    -webkit-appearance: none;
    width: 100%;
    height: 15px;
    border-radius: 5px;
    background: #d3d3d3;
    outline: none;
    opacity: 0.7;
    -webkit-transition: .2s;
    transition: opacity .2s;
    }
    .slider::-webkit-slider-thumb {
    -webkit-appearance: none;
    appearance: none;

    width: 25px;
    height: 25px;
    border-radius: 50%;
    background: #00008B;
    cursor: pointer;
    }
    .slider::-moz-range-thumb {
    width: 25px;
    height: 25px;
    border-radius: 50%;
    background: #4CAF50;
    cursor: pointer;
    }
    </style>
    </head>
    ```

  - Markdown is included to create a div containg the map and a div containing the time slider button:

    ```
    <body>
    <!-- Create map container -->
    <div id="map" class="map"></div>
    <div role="group" aria-label="Animation controls">
    <!-- Create slider container -->
    <div class="slidecontainer">
    <input type="range" min="0" max="2" value="0" class="slider" id="myRange">
    <p>Date: <span id="date_value"></span></p>
    </div>
    ```

  - JavaScript can also be included as a seperate .JS file similar and is added to refer to the WMS and animate the time slider:

    ```
    <script>
    // Set up the layers
    var layers = [
    new ol.layer.Tile({
    source: new ol.source.OSM()
    }),
    new ol.layer.Tile({
    source: new ol.source.TileWMS({
    url: 'http://xx.XXX.XXX.XXX:8080/geoserver/cite/wms',
    params: {'LAYERS': 'cite:timeseries'}
    })
    })
    ];
    var map = new ol.Map({
    layers: layers,
    target: 'map',
    view: new ol.View({
    // Defining the location in Lat Lon.
    center: ol.proj.transform([-120.5542,43.8041], 'EPSG:4326', 'EPSG:3857'),
    zoom: 5
    })
    });
    // Define the available dates
    var dates = ['1981-01-01T00:00:00.000Z', '1991-01-01T00:00:00.000Z',
    '2001-01-01T00:00:00.000Z', '2011-01-01T00:00:00.000Z']
    var sliderRange = document.getElementById("myRange");
    sliderRange.max = dates.length-1;
    var dateValue = document.getElementById("date_value");
    dateValue.innerHTML = dates[sliderRange.value].slice(0,10);
    layers[1].getSource().updateParams({'TIME': dates[sliderRange.value]});
    // Update the current slider value (each time you drag the slider handle)
    sliderRange.oninput = function() {
    dateValue.innerHTML = dates[this.value].slice(0,10);
    layers[1].getSource().updateParams({'TIME': dates[this.value]});
    }
    </script>
    </body>
    </html>
    ```
