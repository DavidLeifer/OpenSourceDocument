1. Highest and Lowest Number Part 1 of 3

  **Week 1 Homework: Highest and Lowest**

    - Make a new class 'Graph' and a function 'hi_lo' that returns the highest number and lowest number from a list of numbers.

    Grading Scale:
    0. Did not attempt
    1. Function calculates the highest number
    2. Function calculates the highest and the lowest number

  - Finding the lowest and highest number is a basic algorithm that can be accomplished in a time complexity of O(n) and auxiliary space O(1).

    - One method is to sort the list and return the first and last value of the list. Time complexity is O(n log n).

    - Another method is linear search, which sets the high and low number and replaces the values by comparison in a loop. Time complexity is 0(n).

    - Other methods include the tournament approach which divides the dataset into two parts but increases auxiliary space to O(log n) or pairwise comparison which is the same speed and space as linear search.
