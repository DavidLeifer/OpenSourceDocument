```python
import sys
import scipy
# print(sys.version)
# Open file function

# Function to open file, 1 to print contents, 0 to not print
def file_opener(data_path, p_bool):
  with open(data_path, "r") as data_open:
    data_read = data_open.read()
    if p_bool == 1:
      print(data_read)
    return data_read
data1, data2, print_cont = "/content/stamina.csv", "/content/food.csv", 0
data1_read, data2_read = file_opener(data1, print_cont), file_opener(data2, print_cont)
# data1_read.close(), data2_read.close()
```

```python
# Index the commas and line breaks, get column length = 1, no = 0
def comma_index(open_file, column_len):
  data_comma_place = []
  column_pl_len = 0
  for i in range(len(open_file)):
    data1_col1 = open_file[i]
    if data1_col1 == ",":
      data_comma_place.append(i)
    if data1_col1 == "\n":
      data_comma_place.append(i)
      # If you want to use the function to get the column length, set to 1
      if column_len == 1:
        data_comma_place = len(data_comma_place)
        break
  return data_comma_place
# Splitting the csv characters into list of words based on indexed comma position
def csv_value_list(data_comma_out, open_file):
  j = 0
  data_val_list = []
  for k in data_comma_out:
    data_val = open_file[j:k]
    data_val_list.append(data_val)
    j = k + 1
  return data_val_list
# Indexed position of the commas in the csv file
data1_comma_place = comma_index(data1_read, 0)
print(data1_comma_place)
# Each value of the csv split into a list based on comma position
data1_comma_list = csv_value_list(data1_comma_place, data1_read)
print(data1_comma_list)
# Length of columns of the csv
data1_col_len = comma_index(data1_read, 1)
print(data1_col_len)
```
