**Homework Notes**

>- The point of this chapter is to use a for loop to automate reading GeoTiffs into Numpy matrices for statistical operations.

>- The first step is to write the iterative loop by accessing the list of GeoTiff strings created in the previous chapter and reading with the GDAL method <i>open</i>.

```python
for input_raster in input_raster_folder_list:
  dataset = gdal.Open(input_raster)
```

>- The next step creates a variable named <i>band</i> to define each of the file's bands with the <i>GetRasterBand</i> function. The example bands are all at the integer 1 but might vary depending on your rasters.

```python
for input_raster in input_raster_folder_list:
  dataset = gdal.Open(input_raster)

  band = dataset.GetRasterBand(1)
```

>- The <i>ReadAsArray</i> method creates a matrix or list of numbers from the GDAL band.

```python
for input_raster in input_raster_folder_list:
  dataset = gdal.Open(input_raster)
  band = dataset.GetRasterBand(1)

  array = band.ReadAsArray()
```

>- No data values are sometimes represented by 0, NAN, or 9999 depending on the GeoTiff.

>- By using a mask with the <i>numpy</i> array, quantiles or numerical breaks can be generated without worrying about calculating strings and integer data types.

>- The <i>GetNoDataValue</i> function defines an array with the file's no data.

```python
for input_raster in input_raster_folder_list:
  dataset = gdal.Open(input_raster)
  band = dataset.GetRasterBand(1)
  array = band.ReadAsArray()

  nodata_val = band.GetNoDataValue()
```

>- An <i>if</i> statement is used to define a masked python list with Numpy's <i>masked_equal</i> method for the array variable containing the GeoTiff's band and the previously defined not a number array <i>nodata_val</i>.

```python
for input_raster in input_raster_folder_list:
  dataset = gdal.Open(input_raster)
  band = dataset.GetRasterBand(1)
  array = band.ReadAsArray()
  nodata_val = band.GetNoDataValue()

  if nodata_val is not None:
    array_mask = np.ma.masked_equal(array, nodata_val)
```

>- The Python list index function is used with the greater than or equal operator and the <i>min</i> function to generate an array that includes the integers but not the strings.

```python
for input_raster in input_raster_folder_list:
  dataset = gdal.Open(input_raster)
  band = dataset.GetRasterBand(1)
  array = band.ReadAsArray()
  nodata_val = band.GetNoDataValue()
  if nodata_val is not None:
    array_mask = np.ma.masked_equal(array, nodata_val)

  array_ignored_nan = array_mask[array_mask >= array_mask.min()]
```

>- The final step is to create a <i>numpy</i> array of 0 integers with the same dimensions as the input array with the <i>zeros_like</i> function. The variable will be used to generate quantiles in later weeks.

```python
for input_raster in input_raster_folder_list:
  dataset = gdal.Open(input_raster)
  band = dataset.GetRasterBand(1)
  array = band.ReadAsArray()
  nodata_val = band.GetNoDataValue()
  if nodata_val is not None:
    array_mask = np.ma.masked_equal(array, nodata_val)
  array_ignored_nan = array_mask[array_mask >= array_mask.min()]

  output = np.zeros_like(array_mask).astype(np.uint8)
```

**Notes**

- Quantiles attempt to set break points along a continuous dataset to visually classify information.

>>- Another way to say this is that quantiles takes the range of the probability distribution and bins them into containers.

>- Quartiles or percentiles are alternative names because they are commonly used.

>>- Quartiles have four groups.

>>- Percentiles have 100 groups. Standardized testing sometimes uses this terminology.

>- The Numpy library has built-in functions to perform these calculations. See the example from their documentation below where 50 is the dividing point.

```python
import numpy as np

a = np.array([[10, 7, 4], [3, 2, 1]])

print(a)
# terminal output: array([[10,  7,  4],
#                         [ 3,  2,  1]])

np.percentile(a, 50)
# terminal output: 3.5
```
