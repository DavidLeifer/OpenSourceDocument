Parametric Vocabulary Review

>- Mean

>>- Greek letter $\mu$

>>- The sum of all numbers divided by the count of values and is used in parametric statistics.

>- Median

>>- The middle most number of the dataset and is typically used in non-paramteric statistics.

>- Population Standard Deviation

>>- In a normal distribution a percentage of the data falls one or more standard deviations from the mean.

$$ \sigma = \sqrt{\frac{\sum_{i=1}^{N} ( x_i - \mu)^2}N } $$

>>- Each value in the dataset ($ \sum_{i=1}^{N} {( x_i ...} $) minus the mean ($ \mu $).

>>- That number is squared.

>>- Then the sum of those numbers is divided by the count of Numbers (N).

>>- The square root of that number is then the population standard deviation.

>>- The other version is sample standard deviation and is nearly the same except it subtracts '1' from the count of Numbers (N-1) in the denominator.

>- A statistician can calculate the skewness and kurtosis which represent the third and the fourth moment using nearly an identical equation.

>- An obsessive statistician can calculate 4-7 or more moments but they are more influenced by the outliers.

**Central Moment Vocabulary**

>- Recall that variance or standard deviation is the second central moment of a probability distribution and is represented by lower case sigma.

>- Skewness - is the third central moment.

>>- There are several versions for different scenarios but the equation below is easier to understand (4).

$$ \gamma_1 = E[(\frac{X - \mu}{\sigma})^3]  $$


>- The same formula can be rewritten (5):

$$ \gamma_1 = \frac{1}{N} \sum_{i=1}^{N} (\frac{X - \mu}{\sigma})^3  $$

>>- The formula subtracts each value in the list ($\sum_{i=1}^{N}(x_i ...$) by the overall mean ($\mu$) and multiplies itself three times or to the third power.

>>- The sum of that number is then calculated.

>>- The sum is divided by the count of values in the list.

>>- That number is divided by the standard deviation to the power of 3 (or cubed).

>- The equation is for the entire population of the data or all the numbers.

>>- In contrast to a sample of the dataset, which subtracts 1 from the count ($\frac{1}{N-1}$) before the first fraction.

>- Kurtosis - UCLA provides definition for three kurtosis equations (6):

>>- Two sums of powered deviation scores, which are used later:

$$s_2 = (x - \bar{x}) ^ 2$$

$$s_4 = (x - \bar{x}) ^ 4$$

>>- Kurtosis is the same formula as skewness but replaces the numerator to the power of 4 instead of 3. It is used in the STATA program:

$$m_4 = \frac{s_4}{N}$$

$$m_2 = \frac{s_2}{N}$$

>>>- The equation is expressed as:

$$ \beta_2 = \frac{m_4}{(m_2)^2} $$

>> The second definition of kurtosis used by the SAS summarize function:

$$ \gamma_2 = \frac{m_4}{(m_2)^2} - 3 $$

>>- The third equation listed by UCLA includes anunbiased estimate:

$$ V(x) = \frac{s_2}{N-1} $$

>- The last formula is overkill for this course but is worth mentioning.

>- It is used in SPSS and cites a more recent textbook.

>- It is noted as being less biased except for non-normal samples or data that is not a symmetrical bell shaped.

```python
# Returns the 2-4 moment of the distribution
def mnt(self, header, mean, col_list):
    col_1 = len(col_list) - 1
    stn = 0
    skew = 0
    kurt = 0
    for i in col_list[1:]:
      n1 = int(i) - mean
      n1_sqr = n1 ** 2
      n1_cube = n1 ** 3
      n1_quad = n1 ** 4
      stn = stn + n1_sqr
      skew = skew + n1_cube
      kurt = kurt + n1_quad
    stn_small_sqr = float(stn) / float(col_1)
    stn_small = stn_small_sqr ** .5
    skew_small_sqr = float(skew) / float(col_1)
    skew_small = skew_small_sqr / (stn_small ** 3)
    kurt_small_sqr = float(kurt) / float(col_1)
    kurt_small = kurt_small_sqr / (stn_small ** 4)
    return [header, stn_small, skew_small, kurt_small]
```

>- Pearson Correlation

>>- Mathematically describes a relationship between two continuous interval variables.
  
>>- Based on assumptions of the Central Limit Theorem on continuous variables.

>>>- A bad example is a person's weight and the price of a coffee thermos.

>>>- Although there may or may not be a mathematical relationship, they are not related.

>>>- It is an indication that further attention should be paid to the variables.

>>- Represented by the lowercase Greek letter ($\rho$) and scales the covariance ($\sigma_{xy}$) between -1 and 1.

>>>- An example of positive correlation that is also cause and effect is income and spending.

>>- Pearson correlation is used to define a relationship that warrants additional investigation when examining larger datasets that otherwise might not be readily apparent.

>- Covariance

>>- The first step in finding Pearson's correlation for a population is to calculate the covariance, which is the numerator of the formula.

$$ \sigma_{xy} = \frac {\sum^N_{i=1} (x_i - \bar x)(y_i - \bar y)}{N} $$

>>- Calculate the first value $x_i$ minus the $x$ mean.

>>- Calculate the second value $y_i$ minus the $y$ mean.

>>- Multiply these two values together.

>>- The final step is to divide by the count ($N$).

>- The result is the covariance, which is an unscaled mathematical relationship (7).

**<u>Week 11. Covariance and Correlation Part 2 of 2: Correlation</u>**

**Week 10 Homework Reminder**

>- Pearson's Correlation Continued

>- The formal and confusing way to express correlation is (7):

$$ \rho = \frac{\sigma_{xy}}{\sigma_x \sigma_y} $$

>>- The beginning of the formula is covariance in the numerator.

>>- The denominator is the standard deviation for $x$ multiplied by the standard deviation for $y$.

>>- The result is a scaled mathematical relationship.

$$ r = \frac{cov(x,y)}{\sigma_x \sigma_y} $$

>>- The numerator and denominator look identical in the first formula.

>>- The second formula explicitly defines the top number as the covariance calculated in Week 10 and the bottom as the product of both column's standard deviation.

**Least Squares Regression Introduction**

>- R squared or coefficient of determination is a measurement to see how well a predictive formula performs.

>- It is used on continuous variables of real measurements like the calorie and alcohol serving data from "B0.csv".

>- This example uses Least squared regression as the predictor line.

>>- The standard slope of a line $y = mx + b$ is similar.

>- Least square regression (8,9)

$$ \hat{y_i} = a + b(x_i - \bar{x}) $$

>- Least squares estimates (8,9):

$$ a = \bar{y} $$

$$ b = \frac{\sum^N_{i=1}(x_i-\bar{x})(y_i-\bar{y})}{\sum^N_{i=1}(x_i-\bar{x})^2} $$

**Week 10 Homework Due**

**Week 10 Homework Example:**

```python
# Covariance and correlation
def covar(self, x_mean, y_mean, col_1_list, col_2_list):
    col_len = len(col_1_list) - 1
    covar = 0
    x1y1_sum = 0
    for i in range(col_len):
      # print("x_mean: ", x_mean[1], "x_value: ", col_1_list[i+1])
      x1 = float(col_1_list[i+1]) - x_mean[1]
      y1 = float(col_2_list[i+1]) - y_mean[1]
      x1y1 = x1 * y1
      x1y1_sum = x1y1_sum + x1y1
    covar = x1y1_sum / col_len
    return covar
def cor(self, covar, col_1_stnd, col_2_stnd):
    stnd12 = col_1_stnd * col_2_stnd
    cor = covar / stnd12
    return cor
```

**Least Squares Regression Continued**

>- The total sum of squares is how much the information varies around the mean (9,10,11).

$$SST0 = \sum^n_{i=1} (y_i - \bar{y})^2$$

>- Regression sum of squares quantifies how far the regression line is from the no relationship line (9,10,11).

$$SSR = \sum^n_{i=1} (\hat{y_i} - \bar{y})^2$$

>- Error sum of squares quantifies how much the data points vary around the estimated regression lines. (9,10,11)

$$SSR = \sum^n_{i=1} (y_i - \hat{y})^2$$
